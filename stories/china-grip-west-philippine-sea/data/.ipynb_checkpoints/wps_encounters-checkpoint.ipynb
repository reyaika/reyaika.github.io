# West Philippine Sea encounters

In this project I would like to map the encounters in the West Philippine Sea, produce static maps which I would then create as a video or a GIF (plays on load). Perhaps create some charts based on the data.

I plan to scrape the posts from the West Philippine Sea Transparency Initiative and categorize them using OpenAI API model to see which posts are possible encounters. From there, I will use the data to tell me which dates I should look at Copernicus and download those to create the image.


Source: 
WPS Transparency Initiative January 2023-February 2025
https://wpstransparency.ph/real-time-tweets/

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import matplotlib.pyplot as plt
```

# Scraping the reports from the West Philippine Sea Transparency Initiative site

```python
wps_url = "https://wpstransparency.ph/real-time-tweets/"
response = requests.get(wps_url)
soup = BeautifulSoup(response.content, 'html.parser')

print(soup.prettify())
```

Look at what the container has

```python
items = soup.select(".elementskit-post-body")
items
```

Scrape:

```python
posts = []

for item in items:
    # Title and URL
    title_tag = item.select_one("h2.entry-title a")
    title = title_tag.text.strip() if title_tag else None
    url = title_tag["href"] if title_tag else None

    #Date
    date_tag = item.select_one("span.meta-date-text")
    date_post = date_tag.text.strip() if date_tag else None

    #Snippet
    snippet_tag = item.select_one("p")
    snippet = snippet_tag.text.strip() if snippet_tag else None

    posts.append({
        "title": title,
        "url": url,
        "date_posted": date_post,
        "snippet": snippet
    })
    
posts[:3]
```

Loop through each URL to get the full text

```python
def get_full_text(url):
    request_full = requests.get(url)
    soup_full = BeautifulSoup(request_full.text, "html.parser")

    paragraphs = soup_full.select(".elementor-widget-container p")

    # Join them together
    text = "\n\n".join([paragraph.get_text(strip=True) for paragraph in paragraphs])
    return text
```

```python
def get_images(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")

    containers = soup.select(".elementor-widget-container")
    img_urls = []

    for container in containers:
        images = container.find_all("img")
        for img in images:
            src = img.get("src") or img.get("data-src")
            if src:
                # Exclude the common header image
                if "West-Philippine-Sea-V3" not in src:
                    img_urls.append(src)

    return img_urls
```

```python
import time

for post in posts:
    post["full_text"] = get_full_text(post["url"])
    post["image_urls"] = get_images(post["url"])
    time.sleep(1)
```

# Classify posts using ChatGPT

I used ChatGPT to classify the incidents according to the kind of encounters. I used the prompt below:

"You are an assistant that classifies reports about West Philippine Sea incidents. create a new  column called ai_guess and classify the following text into one of these categories:
1. "Coast Guard encounter" → The incident mainly involves the Philippine Coast Guard directly interacting or confronting Chinese vessels.
2. "Fisherman encounter" → The incident mainly involves Filipino fishermen or fishing vessels encountering Chinese vessels or aggression.
3. "No encounter" → The text does not describe a direct encounter, but instead general operations, flights, patrols, or policy statements.

Return only one of these labels. Then, send the csv."

Upon download, I checked each row and subsequently created a new column named "groundtruth"

Upon checking the AI answers, I created a new category called "Coast Guard sightings" for incidents that do not necessarily engaged with the Chinese vessels but spotted them.

The accuracy rate of the model was at 58.16%.

I later created a new column called "actual_date" and checked the dates for the ff:
* Coast Guard encounter
* Coast Guard sightings
* Fisherman encounter

Each date was filled in by checking the accompanying image, the text, and the date the tweet was posted if there was a link.

Make actual_date into datetime

```python
def parse_date_range(text):
    if pd.isna(text):
        return pd.Series([pd.NaT, pd.NaT])
    
    text = text.strip()

    # Case 1: Date range like "February 1-9, 2024"
    match_range = re.match(r"([A-Za-z]+) (\d+)-(\d+), (\d{4})", text)
    if match_range:
        month, start_day, end_day, year = match_range.groups()
        start_date = pd.to_datetime(f"{month} {start_day} {year}")
        end_date   = pd.to_datetime(f"{month} {end_day} {year}")
        return pd.Series([start_date, end_date])

    # Case 2: Single date like "March 9, 2023"
    match_single = re.match(r"([A-Za-z]+) (\d+), (\d{4})", text)
    if match_single:
        month, day, year = match_single.groups()
        date = pd.to_datetime(f"{month} {day} {year}")
        return pd.Series([date, date])  # same start & end

    return pd.Series([pd.NaT, pd.NaT])
```

```python
encounters[["start_date", "end_date"]] = encounters["date_posted"].apply(parse_date_range)
```

Find the largest gap (in days) between encounters since 2024

```python
cg_encounters_2024 = coast_guard_encounters[coast_guard_encounters['start_date'].dt.year >= 2024]

cg_encounters_2024.tail()
```

```python
unique_days_2024 = cg_encounters_2024["start_date"].drop_duplicates()
```

```python
unique_days_sorted = unique_days_2024.sort_values().reset_index(drop=True)
```

```python
gaps = unique_days_sorted.diff()

max_gap = gaps.max()
gap_index = gaps.idxmax()
```

```python
date_after_gap = unique_days_sorted[gap_index]
date_before_gap = unique_days_sorted[gap_index - 1]
```

```python
print(f"Largest gap: {max_gap.days} days")
print(f"Gap was between {date_before_gap.date()} and {date_after_gap.date()}")
```
